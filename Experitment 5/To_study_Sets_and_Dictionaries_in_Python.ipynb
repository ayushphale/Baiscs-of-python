```python
#A set is a collection which is unordered, unchangeable, and unindexed. A set is defined using curly brackets. Elements will be shuffled. Set items can appear in a different order every time you use them. It cannot be defined by index or a key. Duplicates are not allowed.
```

```python
#Creation of Set
thisset = {"Prithvi", "Abhilasha", "Shreeya"}
print(thisset)
```

```python
#Duplicates not allowed
thisset = {"Prithvi", "Abhilasha", "Shreeya", "Abhilasha"}
print(thisset)
```

```python
#The values True and 1 are considered the same value in sets
thisset = {"Prithvi", "Abhilasha", "Shreeya", True, 1, 2}
print(thisset)
```

```python
#A set will allow all kind of datatypes, strings, integers and, boolean values:
set1 = {"code", 34, True, 40, "male", 12.4}      #print[set1] will give error
set1
```

```python
#Once a set is created, you cannot replace its items, but you can add new items

thisset = {"Prithvi", "Abhilasha", "Shreeya"}
thisset.add("Prinka")
print(thisset)
```

```python
#Remove Item
thisset = {"Prithvi", "Abhilasha", "Shreeya"}
thisset.remove("Shreeya")
print(thisset)
```

```python
#Set Operations     #At least 2-3 elements in both the sets must be common.
A = {1, 2, 3, 4}
B = {3, 4, 5, 6}

print("Union: ", A|B)
print("Intersection:", A & B)    #Common in both
print("Difference: ", A-B)
print("Difference: ", B-A)
print("Symmetric Difference: ", A^B)   #Not common

```

```python
#Problem Statement 2: Three students choose elective subjects. Find the common subjects chosen by all the students using set operations. Concept used - Set Intersection, Union, etc.

Student1 = {"Physics", "Chemistry", "Maths", "English", "Accounts"}
Student2 = {"Chemistry", "Maths", "French", "German", "Accounts"}
Student3 = {"German", "Sanskrit", "Maths", "Chemistry", "Business Studies", "Accounts"}

print("The common subjects chosen by the students are: ", Student1 & Student2 & Student3)
```

```python
#Problem Statement 3: One set contains students in the Cricket club, another in the Football Club. Find students who are in: -
#1. In both clubs; 2. In only one club. Concept used - Intersection & Symmetric Difference.

Cricket = {"Ansh", "Prithvi", "Chinmay", "Aarav", "Ayush", "Vidhaan", "Soham", "Siddhant"}
Football = {"Ayush", "Prithvi","Aarav", "Yash", "Soham", "Dhairya", "Aksharr"}

print("The students who are in both clubs are: ", Cricket & Football)
print("The students who are in only club are: ", Cricket^Football)
```

```python
#3. A dictionary stores usernames and passwords. Write a Python program to validate user login.

class_dict = {
  "User1": "pass1",
  "User2": "pass2",
  "User3": "pass3",
}
username = input("Enter username:")
password = input("Enter password:")
if class_dict.get(username) == password:
    print("Login Successful")
else:
      print("Invalid Password")

```

```python
#A dictionary stores student names and marks. Write a PYTHON progra to find the student who scored the highest marks.

marks = {"A": 85, "B": 92, "C": 78}
topper = max(marks, key=marks.get)
print("Topper:", topper)
print("Marks:", marks[topper])
```
